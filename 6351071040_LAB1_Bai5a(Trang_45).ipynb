1. Đọc dữ liệu vào chương trình

```python
import pandas as pd

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('/content/Online_retail.xlsx')

# Display the first 5 rows of the DataFrame
print("Huỳnh Thị Trúc Lam - 6351071040")
df.head()
```

2. Tiền xử lý dữ liệu: cắt bỏ các ký tự thừa ở tên mặt hàng mua (cột Description), xóa các dòng dữ liệu không có số hóa đơn (cột InvoiceNo) và chuyển nó về kiểu dữ liệu chuỗi.

```python
df['Description'] = df['Description'].str.strip()
df.dropna(subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype(str)

print("Huỳnh Thị Trúc Lam - 6351071040")
print("DataFrame after cleaning 'Description', dropping NaNs in 'InvoiceNo', and converting 'InvoiceNo' to string:")
df.head()
df.info()
```

3. Trong dữ liệu đã cho, có một số hóa đơn là hóa đơn tín dụng thay vì là hóa đơn ghi nợ vì vậy hãy xóa những hóa đơn đó. Chúng được xác định với ký tự ‘C’ chứa trong số hóa đơn InvoiceNo.

```python
df['Description'] = df['Description'].str.strip()
df.dropna(subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype(str)

print("Huỳnh Thị Trúc Lam - 6351071040")
print("DataFrame after cleaning 'Description', dropping NaNs in 'InvoiceNo', and converting 'InvoiceNo' to string:")
df.head()
df.info()
```

```python
df = df[~df['InvoiceNo'].astype(str).str.contains('C', na=False)]

print("Huỳnh Thị Trúc Lam - 6351071040")
print("DataFrame after removing credit invoices (InvoiceNo starting with 'C'):")
df.head()
print("Shape of the DataFrame after removing credit invoices:", df.shape)
```

4. Thống kê số dòng dữ liệu theo từng quốc gia.

```python
country_counts = df['Country'].value_counts()
print("Huỳnh Thị Trúc Lam - 6351071040")
print("Number of data rows per country:")
print(country_counts)
```

5. Lấy ra dữ liệu hóa đơn từ nước Anh ‘Germany’ và gom nhóm cột Số lượng
mua (Quantity) theo Số hóa đơn (InvoiceNo) và Tên mặt hàng (Description).

```python
df_uk = df[df['Country'] == 'Germany']
uk_invoice_product_quantity = df_uk.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().reset_index()
print("Huỳnh Thị Trúc Lam - 6351071040")
print("First 5 rows of grouped UK invoice data by product quantity:")
print(uk_invoice_product_quantity.head())
```

6. Chuyển đổi dữ liệu về dạng hot encoding, với mỗi dòng dữ liệu là một hóa đơn.

```python
uk_invoice_product_one_hot = uk_invoice_product_quantity.pivot_table(index='InvoiceNo', columns='Description', values='Quantity').fillna(0)
print("Huỳnh Thị Trúc Lam - 6351071040")
print("First 5 rows of the one-hot encoded UK invoice data:")
print(uk_invoice_product_one_hot.head())
```
